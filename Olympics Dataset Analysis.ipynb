```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
athletes = pd.read_csv('C:/datasets/olympic dataset/athlete_events.csv')
regions = pd.read_csv('C:/datasets/olympic dataset/noc_regions.csv')
```

```python
top_10_countries = athletes_df.Team.value_counts().sort_values(ascending=False).to_frame().head(10)
top_10_countries
```

```python
plt.figure(figsize=(12,6))
plt.title('Overall participation by country')
sns.barplot(data=top_10_countries, x= top_10_countries.index, y=top_10_countries.Team, palette= 'Set2')
plt.xlabel('Teams')
plt.ylabel('Count')
plt.show();
```

```python
plt.figure(figsize=(12,6))
plt.title('Age distribution Of the athletes')
plt.xlabel('Age')
plt.ylabel('Number of participants')
plt.hist(athletes_df.Age, bins= np.arange(10,80,2), color='orange', edgecolor='white');
```

```python
#Total number of female athletes in each olmypics
female_participants = athletes_df[(athletes_df.Sex=='F') & (athletes_df.Season=='Summer')][['Sex','Year']]
female_participants = female_participants.groupby('Year').count().reset_index()
female_participants.tail()
```

```python
womenOlympics = athletes_df[(athletes_df.Sex == 'F')  & (athletes_df.Season=='Summer')]
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(20,10))
sns.countplot(x='Year', data=womenOlympics, palette='tab10')
plt.title('Women participation')
```

```python
#female Arhletes over Time
part = womenOlympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20,10))
part.loc[:,"F"].plot()
plt.title('Plot Of female Arhletes over Time')
```

```python
goldMedals = athletes_df[athletes_df.Medal == 'Gold']
goldMedals.head()
```

```python
# take only values that are different from nan.
goldMedals = goldMedals[np.isfinite(goldMedals['Age'])]
 
```

```python
# Gold Beyond 60
goldMedals['ID'][goldMedals['Age'] > 60].count()
```

```python
sporting_event = goldMedals['Sport'][goldMedals['Age'] > 60]
sporting_event
```

```python
plt.figure(figsize=(10,5))
plt.tight_layout()
sns.countplot(sporting_event)
plt.title('Gold medal for athletes over 60 years')
```

```python
# Gold medal from each country.

goldMedals.Region.value_counts().reset_index(name='Medal').head()
```

```python
totalGoldMedals = goldMedals.Region.value_counts().reset_index(name='Medal').head(6)
g = sns.catplot(x='index', y='Medal',data=totalGoldMedals, kind='bar',palette='rocket')
#g.despine(right=True)
g.set_xlabels('Top 5 Countries')
g.set_ylabels('Number of Medals')
plt.title('Gold meadals per country')
```

```python
max_year = athletes_df.Year.max()
#print(max_year)

team_names = athletes_df[(athletes_df.Year == max_year) & (athletes_df.Medal == 'Gold')]['Team']
team_names.value_counts().head(10)
```

```python
sns.barplot(x= team_names.value_counts().head(20), y= team_names.value_counts().head(20).index)
plt.ylabel(None)
plt.xlabel('countrywise medals for the year 2016')
```
